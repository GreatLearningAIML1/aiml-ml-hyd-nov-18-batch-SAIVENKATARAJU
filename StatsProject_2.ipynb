```python
import pandas as pd
import numpy as np
```

```python
Sales_data=pd.read_excel('Sales.xlsx')
```

```python
Old=Sales_data['Old Scheme (in thousands)']
New=Sales_data['New Scheme (in thousands)']
```

```python
np.mean(Old)
```

```python
np.mean(New)
```

# Performing one tailed test

```python
from scipy.stats import ttest_ind,ttest_1samp
```

```python
ttest_ind(Old,New)
```

#  2)Since p<0.5 we failed to reject null hypothesis i,e  new scheme has significantly NOT raised output

# 3)Probability of Type I error = significant level = 0.05 or 5%

```python
t,p=ttest_1samp(New-Old,5)
print(p)
```

# Since it is a one sided test Pvalue 0.700 which is greater than level of significance Î± (0.05) , hence we retain the null hypothesis or we fail to reject the Null Hypothesis on
