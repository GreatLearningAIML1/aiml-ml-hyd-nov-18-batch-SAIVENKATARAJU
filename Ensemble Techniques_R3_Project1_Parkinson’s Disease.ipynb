```python
# Removing HNR as less co realtion
```

```python
Feature=Pd_df.drop(columns=['name','status','HNR'])
```

```python
Target=Pd_df.pop('status')
```

```python
from sklearn.model_selection import train_test_split

Xtrain,Xtest,Ytrain,Ytest=train_test_split(Feature,Target,test_size=0.30,random_state=1)
```

```python
Xtrain.shape,Ytrain.shape,Xtest.shape,Ytest.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='entropy')
```

```python
model.fit(Xtrain,Ytrain)
```

```python
from sklearn.metrics import accuracy_score,precision_score
Ypred=model.predict(Xtest)
```

```python
accuracy_score(Ytest,Ypred)
```

```python
precision_score(Ytest,Ypred)
```

```python
Ypred.shape,Ytest.shape
```

```python
pd.crosstab(Ytest,Ypred,colnames=['predicted'])
```

```python
# increase the max depth and min_sample_leaf
model_w_reg=DecisionTreeClassifier(criterion='entropy',max_depth=15,min_samples_leaf=2,random_state=1)
```

```python
model_w_reg.fit(Xtrain,Ytrain)
```

```python
Ypred_w_reg=model_w_reg.predict(Xtest)
```

```python
accuracy_score(Ytest,Ypred_w_reg)
```

```python
precision_score(Ytest,Ypred)
```

```python
#with the increase of min_sample_leaf from 1 to 2 accuracy has increased but in general it cause underfit the data as less data in every node
```

```python
#with change of Max_dept to 15 accuracy increased but in general it cause model to be overfitted
```

```python
pd.crosstab(Ytest,Ypred_w_reg,colnames=['predicted'])
```

```python
from sklearn.ensemble import RandomForestClassifier
model_rt=RandomForestClassifier(n_estimators=15,random_state=1)
model_rt.fit(Xtrain,Ytrain)
```

```python
Ypred_rt=model_rt.predict(Xtest)
```

```python
accuracy_score(Ytest,Ypred_rt)
```

```python
precision_score(Ytest,Ypred_rt)
```

```python
pd.crosstab(Ytest,Ypred_rt,colnames=['predicted'])
```

```python
from sklearn.model_selection import GridSearchCV
model_rt_gc=RandomForestClassifier(random_state=1)
grid_param = {  
    'n_estimators':list(range(1,16)),
    'criterion': ['gini', 'entropy'],
    'bootstrap': [True, False]
}
```

```python
gr_s=GridSearchCV(estimator=model_rt_gc,param_grid=grid_param,scoring='accuracy',cv=5,n_jobs=1)
```

```python
gr_s.fit(Xtrain,Ytrain)
```

```python
gr_s.best_params_
```

```python
model_rt_after_gc=RandomForestClassifier(random_state=1,**gr_s.best_params_)
```

```python
model_rt_after_gc.fit(Xtrain,Ytrain)
```

```python
Ypred_rt_gc=model_rt_after_gc.predict(Xtest)
```

```python
accuracy_score(Ytest,Ypred_rt_gc)
```

```python
precision_score(Ytest,Ypred_rt_gc)
```
